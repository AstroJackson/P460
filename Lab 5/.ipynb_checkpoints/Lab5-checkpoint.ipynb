```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, matplotlib as mpl
from scipy.optimize import curve_fit
```

```python
data = pd.read_excel("Lab 5 Data.xlsx", sheet_name = f"2 Polarizers")

angle = data["theta"] * np.pi / 180
intensity = data["I"]

def I2(theta, a, b): #intensity with two polarizers
    return a + b * (np.cos(theta))**2

initialParameters = [-6, 30]

finalParameters, parCov = curve_fit(I2, angle, intensity, sigma = None, p0=initialParameters, absolute_sigma=False)
a = finalParameters[0]
b = finalParameters[1]
sigma_a = parCov[0][0]
sigma_b = parCov[1][1]

#a, b = -6, 30

synth_angle = np.linspace(min(angle), max(angle), 1000)
synth_intensity = I2(synth_angle, a, b)

plt.figure(figsize = (15, 8))
plt.title("Intensity vs. Angle - Two Polarizers")
plt.plot(angle*180/np.pi, intensity, 'o', label = "Data")
plt.plot(synth_angle*180/np.pi, synth_intensity, label = f"Theory with a = {round(a, 2)}\
 +- {round(sigma_a, 2)} and b = {round(b,2)} +- {round(sigma_b, 2)}")
plt.xlabel("Angle (degrees)"); plt.ylabel("Intensity")
plt.legend(loc = "upper right")
plt.show()

#print(parCov)
```

```python
data = pd.read_excel("Lab 5 Data.xlsx", sheet_name = f"3 Polarizers")

angle = data["theta"] * np.pi / 180
intensity = data["I"]

def I3(theta, a, b): #intensity with three polarizers
    return a + b * (np.sin(2*theta))**2

initialParameters = [0, 3]

finalParameters, parCov = curve_fit(I3, angle, intensity, sigma = None, p0=initialParameters, absolute_sigma=False)
a = finalParameters[0]
b = finalParameters[1]
sigma_a = parCov[0][0]
sigma_b = parCov[1][1]

#a, b = 0, 3

synth_angle = np.linspace(min(angle), max(angle), 1000)
synth_intensity = I3(synth_angle, a, b)

plt.figure(figsize = (15, 8))
plt.title("Intensity vs. Angle - Three Polarizers")
plt.plot(angle*180/np.pi, intensity, 'o', label = "Data")
plt.plot(synth_angle*180/np.pi, synth_intensity, label = f"Theory with a = {round(a, 2)} \
+- {round(sigma_a, 2)} and b = {round(b,2)} +- {round(sigma_b, 2)}")
plt.xlabel("Angle (degrees)"); plt.ylabel("Intensity")
plt.legend(loc = "upper right")
plt.show()
```
