```python
import numpy as np, pandas as pd, matplotlib as mpl, matplotlib.pyplot as plt, os
from scipy.optimize import curve_fit

os.chdir("/Users/jackson/Desktop/P460/Lab 1")

data = pd.read_excel("P460Lab1.xlsx", sheet_name="Data")
D, sigma_D = data["D"], data["dD"]
s01, sigma_s01 = data["S01"], data["dS01"]
s02, sigma_s02 = data["S02"], data["dS02"]
b1, sigma_b1 = data["b1"], data["db1"]
b2, sigma_b2 = data["b2"], data["db2"]

si1 = D - s01
sigma_si1 = np.sqrt(sigma_D**2 + sigma_s01**2)
si2 = D - s02
sigma_si2 = np.sqrt(sigma_D**2 + sigma_s02**2)

f_known = 12.7
object_size = 0.8
```

```python
def s0_plus(D, f):
    return D/2 * (1 + np.sqrt(1 - 4 * f / D))
def s0_minus(D, f):
    return D/2 * (1 - np.sqrt(1 - 4 * f / D))
D_synthetic = np.linspace(D[0], D[len(D)-1], 50)
s0_synthetic_plus = [s0_plus(D = D_synthetic[i], f= 12.7) for i in range(len(D_synthetic))]
s0_synthetic_minus = [s0_minus(D = D_synthetic[i], f = 12.7) for i in range(len(D_synthetic))]

plt.figure(figsize = (10,8))
plt.errorbar(D, s01, sigma_s01, fmt = 'o', label = "s01", markersize = 5)
plt.errorbar(D, s02, sigma_s02, fmt = 'go', label = "s02", markersize = 5)

initialParameters = 12.7
finalParameters, parCov = curve_fit(s0_plus, D, s01, sigma = sigma_s01, p0=initialParameters, absolute_sigma=False)
best_f1 = finalParameters[0]
sigma_best_f1 = parCov[0][0]

initialParameters = 12.7
finalParameters, parCov = curve_fit(s0_minus, D, s02, sigma= sigma_s02, p0=initialParameters, absolute_sigma=False)
best_f2 = finalParameters[0]
sigma_best_f2 = parCov[0][0]

plt.plot(D_synthetic, s0_synthetic_plus, color = "orange", label = f"f = {round(best_f1, 1)} cm")
plt.plot(D_synthetic, s0_synthetic_minus, color = "purple", label = f"f = {round(best_f2, 1)} cm")

colors = ["black", "crimson"]
for i, f in enumerate([best_f1-.7, best_f1+.7]):
    s0_synthetic_plus = [s0_plus(D = D_synthetic[i], f = f) for i in range(len(D_synthetic))]
    plt.plot(D_synthetic, s0_synthetic_plus, color = colors[i]) #label = f"f = {round(f, 3)} cm"
    

colors1 = ["skyblue", "yellow"]
for i, f in enumerate([best_f2-.7, best_f2+.7]):
    s0_synthetic_minus = [s0_minus(D = D_synthetic[i], f = f) for i in range(len(D_synthetic))]
    plt.plot(D_synthetic, s0_synthetic_minus, color = colors1[i])

plt.xlabel("D (cm)"); plt.ylabel("s0 (cm)")
plt.legend()
plt.title("D vs. s0 fit using the Levenberg-Marquardt algorithm")
plt.figtext(.3, .85, f"Best Focal Length for s01 = {round(best_f1,1)} +- 0.1 cm",
            fontweight = "roman", fontsize = 12)
plt.figtext(.3, .83, f"Best Focal Length for s02 = {round(best_f2,1)} +- 0.1 cm",
            fontweight = "roman", fontsize = 12)
plt.figtext(.45, .3, "(Less accurate focal lengths shown as well)",
            fontweight = "roman", fontsize = 12)
plt.show()
```

```python
# weight1 = 1 / (sigma_best_f1**2)
# weight2 = 1 / (sigma_best_f2**2)
weight1 = 1/.1**2
weight2 = 1/.1**2
best_f = (best_f1*weight1 + best_f2*weight2) / (weight1 + weight2)
sigma_best_f = np.sqrt(1 / (weight1 + weight2))
print(f"Best f = {best_f} +- {sigma_best_f} cm")
```

```python
dfds01 = (1/s01 + 1/si1)**-2 * 1/s01**2
dfds02 = (1/s02 + 1/si2)**-2 * 1/s02**2
dfdsi1 = (1/s01 + 1/si1)**-2 * 1/si1**2
dfdsi2 = (1/s02 + 1/si2)**-2 * 1/si2**2

df1 = np.sqrt(dfds01**2 * sigma_s01**2 + dfdsi1**2 * sigma_si1**2)
df2 = np.sqrt(dfds02**2 * sigma_s02**2 + dfdsi2**2 * sigma_si2**2)

f1 = (1/s01 + 1/si1)**-1
f2 = (1/s02 + 1/si2)**-1

weight1 = 1/df1**2
weight2 = 1/df2**2

best_f1 = np.sum(f1*weight1)/ (np.sum(weight1))
best_f2 = np.sum(f2*weight2)/ (np.sum(weight2))
best_f = (np.sum(f1*weight1)+np.sum(f2*weight2)) / (np.sum(weight1) + np.sum(weight2))
df_best = np.sqrt(1 / (np.sum(weight1) + np.sum(weight2)))
df1_best = np.sqrt(1 / (np.sum(weight1)))
df2_best = np.sqrt(1 / (np.sum(weight2)))


print(f"Best f = {best_f} +- {df_best} cm")
print(f"Best f1 = {best_f1} +- {df1_best} cm")
print(f"Best f2 = {best_f2} +- {df2_best} cm")
```

```python
m_calc1 = b1/object_size
m_calc2 = b2/object_size
m_theory1 = si1/s01
m_theory2 = si2/s02

def linear(x, m):
    return m*x
m_t_both = np.concatenate((m_theory1, m_theory2))
m_c_both = np.concatenate((m_calc1, m_calc2))
initialParameters = 1
finalParameters, parCov = curve_fit(linear, m_t_both, m_c_both)#, sigma= sigma_s02, p0=initialParameters, absolute_sigma=False)
best_slope = finalParameters[0]
slope_error = parCov[0][0]
print(f"Best slope = {best_slope} +- {slope_error}")



plt.figure(figsize = (10,6))
plt.plot(m_theory1, m_calc1, 'o', label = "From s01 Data")
plt.plot(m_theory2, m_calc2, 'o', label = "From s02 Data")
plt.plot([0,3.5], [0, 3.5], label = "M_calculated = M_theory line")
plt.plot([0,3.5], [0, 3.5*best_slope], label = f"Line of best fit (LM Algorithm): y = {round(best_slope, 3)}x"
         , color = "purple")

plt.figtext(.37, .42, f"Line of best fit used both S01 and S02 data.",
            fontweight = "roman", fontsize = 12, rotation = 30.5)
plt.xlabel("-1 * M_theory"); plt.ylabel("-1 * M_calculated")
plt.title("M_calculated vs. M_theory")
plt.legend()
plt.show()
```
