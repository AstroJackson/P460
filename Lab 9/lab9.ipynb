```python
import pandas as pd, matplotlib.pyplot as plt, matplotlib as mpl, numpy as np
from scipy.optimize import curve_fit
from scipy.signal import find_peaks
from scipy.interpolate import interp1d
```

```python
name = "He"
data = pd.read_excel("lab9data.xlsx", sheet_name = f"{name}")
wl = data["wavelength"]
inte = data["intensity"]
fig, ax = plt.subplots(figsize = (15,8))
p1, = ax.plot(wl, inte)
ax.set_xlabel("Wavelength (nm)")
ax.set_ylabel("Intensity (relative units)")
lines = np.array([[587.6, 500], [667.8, 200], [706.5, 100], [728.1, 50], [1083, 1000]])
ax.plot(lines[:,0], lines[:,1]*5, 'o')
for list in lines:
    ax.plot([list[0], list[0]], [0, list[1]*5], linewidth = 1)

ax.set_xlim(min(wl)-2.5, max(wl)+2.5)
ax.set_ylim(0, max(inte)+75)
ax.set_title(f"{name} Lines")
plt.savefig(f"{name}.jpg")
```

```python
name = "tungsten"
data = pd.read_excel("lab9data.xlsx", sheet_name = f"{name}")
wl = data["wavelengthd"]
inte = data["intensityd"]
wlg, integ = data["green"], data["gw"]
wlr, inter = data["red"], data["rw"]
wlir, inteir = data["IRcutoff"], data["irw"]
fig, ax = plt.subplots(figsize = (15,8))
p1, = ax.plot(wl, inte, label = "No Filter")
p2, = ax.plot(wlg, integ, label = "Green Filter")
p3, = ax.plot(wlr, inter, label = "Red Filter")
p4, = ax.plot(wlir, inteir, label = "Infrared Cutoff Filter")

ax.set_xlabel("Wavelength (nm)")
ax.set_ylabel("Intensity (different relative units for each filter)")
plt.legend()
ax.set_xlim(min(wl)-2.5, max(wl)+2.5)
ax.set_ylim(0, max(inte)+75)
ax.set_title(f"Tungsten Lines ")
plt.savefig(f"{name}.jpg")
```
